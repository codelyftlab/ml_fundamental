```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# read the csv file #
df = pd.read_csv('data/salaries-2023.csv')
df.head()
```

```python
# remove the 'NAN' values in our data #
df = df.dropna()
df.head()
```

```python
print(df.shape)
df.info()
```

```python
print(df.describe())
```

```python
df.language.value_counts()
```

```python
# filter data only top 4 languages to use #
languages = ['php', 'js', '.net', 'java']
df = df[df['language'].isin(languages)]
df.shape
```

```python
# filter cities #
df.city.value_counts()
```

```python
vilnius_names = ['Vilniuj', 'Vilniua', 'VILNIUJE', 'VILNIUS', 'vilnius', 'Vilniuje']
condition = df['city'].isin(vilnius_names)
df.loc[condition, 'city'] = 'Vilnius'

kaunas_names = ['KAUNAS', 'kaunas', 'Kaune']
condifiton = df['city'].isin(kaunas_names)
df.loc[condition, 'city'] = 'Kaunas'

df.city.value_counts()
```

```python
# drop all cities and left with the top 2 cities #
cities = ['Vilnius', 'Kaunas']
df = df[df['city'].isin(cities)]
df.shape
```

```python
# sort by highest salary #
df_sorted = df.sort_values(by='salary', ascending=False)
df_sorted.head(30)
```

```python
x = df.iloc[:, -2:-1]
y = df.iloc[:, -1].values

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(x, y)
plt.show()
```

```python
df = df[df['salary'] <= 6000]
df.shape
```

```python
x = df.iloc[:, -2]
y = df.iloc[:, -1].values

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(x, y)
plt.show()
```

```python
one_hot = pd.get_dummies(df['language'], prefix='lang')
df = df.join(one_hot)
df = df.drop('language', axis=1)

one_hot = pd.get_dummies(df['city'], prefix='city')
df = df.join(one_hot)
df = df.drop('city', axis=1)

df.head(20)
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# take only years and level for x #
# take only salary for y #

x = df.iloc[:,0:2].values
y = df.iloc[:,2].values
print(x[0:10])
```

```python
print(y[0:10])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
salaries = model.predict([[3, 12], [1, 1]])
print(salaries)
```

```python
r2 = r2_score(y_test, y_pred)
print(f"R2 Score : {r2} ({r2:.3%})")
```
